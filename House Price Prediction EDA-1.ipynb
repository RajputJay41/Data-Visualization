# Missing Values

```python
## Here we will check the percentage of nan values present in each feature
## 1 -step make the list of features which has missing values
features_with_na=[features for features in dataset.columns if dataset[features].isnull().sum()>1]
## 2- step print the feature name and the percentage of missing values

for feature in features_with_na:
    print(feature, np.round(dataset[feature].isnull().mean(), 4),  ' % missing values')
```

# since there are many missing values , we need to find the relationship between missing values and sales price

```python
for feature in features_with_na:
    data = dataset.copy()
    
    # let's make a variable that indicates 1 if the observation was missing or zero otherwise
    data[feature] = np.where(data[feature].isnull(), 1, 0)
    
    # let's calculate the mean SalePrice where the information is missing or present
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

Here with the relation between the missing values and the dependent variable is clearly visible. so we need to replace these nan values with something meaningful which we will do in feature enginnering section.

# Temporal Variables [Eg:- Date Time Variables]

From the dataset we have 4 year variables. we have to extract the information from the datetime varibales like no of years
or no of days. one example in this specific scenario we can see that there is differnce in years between years of the
house was built and when the house was sold. We will be performing analysis in the feature engineering.

```python
# list of variables that contain year information
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]

year_feature
```

```python
# let's explore the content of these year variables
for feature in year_feature:
    print(feature, dataset[feature].unique())
```

```python
## lets analyze the temporal Datetime variables
# we will check whether there is a relation between the year house is sold and sales price

dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title("House Price vs YearSold")
```

```python
## here we will compare the differnce between All year Feature with sales Price

for feature in year_feature:
    if feature != 'Yrsold':
        data = dataset.copy()
        ### we will capture the differnce between year variable and year the house was sold for
        data[feature] = data['YrSold']-data[feature]
        
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
    
```

```python
## Numerical varaibles are usually of 2 types
## continous  variable and Discrete variables

discrete_feature=[feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_feature+['Id']]
print("discrete Variables Count: {}".format(len(discrete_feature)))
```

```python
discrete_feature
```

```python
dataset[discrete_feature].head()
```

```python
## lets find the relationship between them and sale Price

for feature in discrete_feature:
    data=dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('Saleprice')
    plt.title(feature)
    plt.show()
```

# Continous Variable

```python
continous_feature = [feature for feature in numerical_features if feature not in discrete_feature + year_feature+['Id']]
print("continous Feature Count: {}".format(len(continous_feature)))
```

```python
## analysing the continous variable by creating the histograms

for feature in continous_feature:
    data=dataset.copy()
    data[feature].hist(bins=35)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```
